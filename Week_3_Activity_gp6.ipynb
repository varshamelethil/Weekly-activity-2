### 4.Create a histogram of "Age 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,9))
plt.hist(dataset['Age'],color='red',rwidth=0.6)
plt.title ('Age of patients vs Count',fontsize=18)
plt.xlabel('Age of patients')
plt.ylabel('Count')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(9,9))
plt.scatter(dataset['BMI'],dataset['Glucose'],color='blue')
plt.title('Scatter plot of BMI vs Glucose',fontsize=18)
plt.xlabel('BMI')
plt.ylabel('Glucose')
```
